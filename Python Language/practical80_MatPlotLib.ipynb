```python
# Perform following operations in Matplotlib:

# Q-1 --> 1. Plot a line between (10,12) to (15,80) with following properties:
# line type = dashed dot
# color = red
# line width = 5.6.
# Title of graph = "Line Graph 1"
# X- axis label = X - values
# Y - axis label = Y-values

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([10,12])
ypoints = np.array([15,80])
plt.plot(xpoints, ypoints, linestyle = '-.', color = 'r', linewidth = 5.6)
plt.title("Line Graph 1")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()
```

```python
# Q-2 --> Plot a line with values (12,34,56,60) with line width = 2.5 and title of graph as "Line Graph 2".

ypoints = np.array([12,34,56,60])
plt.plot(ypoints, linewidth = 2.5)
plt.title("Line Graph 2")
plt.show()


```

```python
# Q-3 --> Plot a bar chart for following properties:
# X - Values: Multiples of 5 in the range 10 to 41
# Y - Values: Multiples of 10 in the range 20 to 87
# color : #4CAF50
# width : 0.5

X = np.arange(10,41,5)
Y = np.arange(20,87,10)
plt.bar(X, Y, color = '#4CAF50', width = 0.5)
plt.show()
```

```python
# Q-4 --> Plot a bar chart for following properties:
# X - Values: Multiples of 2 in the range 10 to 23
# Y - Values: Multiples of 10 in the range 20 to 87
# color : #4CAF50
# height : 0.2

X = np.arange(10,23,2)
Y = np.arange(20,87,10)
plt.bar(X, Y, color = '#4CAF50', width = 0.2)
plt.show()
```

```python
# Q-5 --> Plot a pie chart for following data:
# data = [50, 20, 10,15, 5]
# label list = ["C", "Python", "PHP", "Java","Ruby"]
# color list = ["black", "hotpink", "b", "#4CAF50","r"]

y = np.array([20,20,10,15,5])
mylabels = ["C", "Python", "PHP", "Java","Ruby"]
mycolors = ["black", "hotpink", "b", "#4CAF50","r"]
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```
