```python
# Q-1 --> Write a program to add, subtract, multiple and divide two Pandas series.
import pandas as pd

s1 = pd.Series([11,12,13])
s2 = pd.Series([1,2,3])

s3 = s1 + s2
print("\n Sum = \n", s3)

s4 = s1 - s2
print("\n Subtract = \n", s4)

s5 = s1 * s2
print("\n Multiply = \n", s5)

s6 = s1 / s2
print("\n Divide = \n", s6)
```

```python
# Q-2 --> Write a program to convert a dictionary to a Pandas series.

import pandas as pd
dic = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
s = pd.Series(dic)
print(s)
```

```python
# Q-3 --> Write a program to convert a NumPy array to a Pandas series

import pandas as pd
import numpy as np

l1 = [22,23,24]
a = np.array(l1)
s = pd.Series(a)
print(s)
```

```python
# Q-4 --> Write a program to add some data to an existing series.

s1 = pd.Series([11,12,13])
s2 = pd.Series([1,2,3])
s = s1 + s2
print("\n Sum = \n", s)
s3 = s.append(s2)
print("Appended Series = \n", s3)

```

```python
# Q-5 --> Write a program to create and display a DataFrame from a specified dictionary data which has the index labels. Get and print the first 3 rows of the DataFrame.

import pandas as pd
import numpy as np

data = {'Name' :['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'Score' :[12.5, 9, 16.5,np.nan,9, 20, 14.5, np.nan, 8, 19],
        'Attempts' : [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'Qualify' : ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
label = ['a','b','c','d','e','f','g','h','i','j']

s1 = pd.DataFrame(data, index = label)
print("Data with labels : \n")
print(s1)
print()

s2 = pd.DataFrame(data, index = range(1,11))
print(s2)
print()

print("Data of first 3 rows are : \n")
print(s2.iloc[:3])
print() 
```
